```python
'''
1) Data Deletion -> pop( index ), del
2) Data Insertion   -> a) Direct Method,  b) join()
3) Slicing of DataFrames
'''
import pandas as pd
```

```python
pds1 = pd.Series( data = list( range(1, 10, 2) ), index = ['one', 'three', 'five', 'seven', 'nine'] )

pds2 = pd.Series( data = list( range(10, 20, 2) ), index = [ 'ten', 'twelve', 'forteen', 'sixteen', 'eighteen' ] )

pds3 = pd.Series( data = list( range(20, 30, 2) ), index = ['twenty', 'twentytwo', 'twentyfour', 'twentysix', 'twentyeight' ] )

pds3 = pd.Series( data = list( range(20, 30, 2) ), index = ['twenty', 'twentytwo', 'twentyfour', 'twentysix', 'twentyeight' ] )

pds5 = pd.Series( data = list( range(10, 20, 1) ), index = [ 'ten', 'eleven', 'twelve', 'thirteen', 'forteen', 'fifteen', 'sixteen', 'seventeen', 'eighteen', 'nineteen' ] )

pds6 = pd.Series( data = list( range(20, 30, 1) ), index = ['twenty', 'twentyone', 'twentytwo', 'twentythree', 'twentyfour', 'twentyfive', 'twentysix', 'twentyseven', 'twentyeight',  'twentynine' ] )

d = { 'Series1' : pds1, 'Series2' : pds2, 'Series3' : pds3, 'Series4' : pds4, 'Series5' : pds5, 'Series6' : pds6  }
pddf = pd.DataFrame( d )
```

```python
print( pddf )
```

```python
#Data Deletion

s6 = pddf.pop('Series6')     #pop will do the action, Update the DataFrame and returns value
print( s6 )
```

```python
print( pddf )
```

```python
del pddf['Series5']   #del will do the action, Update the DataFrame and does not return any value
```

```python
print( pddf )
```

```python
#Data Insertion - we are going to insert series1 data + 2.

print( pddf['Series1'] )
```

```python
'''
There are 3 method to add data in existing DataFrame.
1) Direct Method, -> dataframe[ key/ new_col ] = panda Series

pddf['Series5'] = pddf['Series1'] + 2
'''
pddf['Series5'] = pddf['Series1'] + 2


'''
2) insert method - location + key(col_name) + data( panda Series )

Syntax :- dataframe.insert( location, 'key', panda series )


print( f"type( pddf['Series1'] ) = {type( pddf['Series1'] )}" )
'''
pddf.insert(2, 'Series6', pddf['Series1'] + 4)
```

```python
pddf
```

```python
'''
3) join method - ( for join by default the location is last position in dataset )

DataFrame.append( Pandas Series )

Syntax :- dataframe.append( key, data )  -> Wrong. Append doesn't work with Panda Series.

DataFrame can only append other DataFrame not a Pandas Series.

Our Approach,

Step1 :- Data -> pddf['Series1']
Step2 :- Key -> 'Series7'

Step3 :- Dictionary( key : Data ) -> {'Series7' : pddf['Series1'] + 6 } -> Now this is Dictionary

Step 4 :- pd.DataFrame( Dictionary ) -> pd.DataFrame( 'Series7' : pddf['Series1'] + 6 ) ) -> Type Casting the Dictionary to make it Panda DataFrame.

Step5 :- join the dataFrame with another DataFrame. -> DataFrame1.join( DataFrame2 )

'''
```

```python
Data = pddf['Series1']+6
print(Data)
print()
key = 'Series7'

d = { key : Data}
print(d)
print( f"type(d) = { type(d) }" )
print()
temp = pd.DataFrame( d )
print( f"type( temp ) = { type( temp ) }" )
pddf.join( temp )   #join() will do the action but not update pddf. So, we will have to save the action in pddf.
```

```python
pddf = pddf.join( temp )
pddf
```

```python
pddf
```

```python
print( pddf.head() )
#join() is like a concatination while adding DataFrames together.
```

```python
#Slicing of DataFrame

print( pddf[ 'eight' : 'five' ] )
print()
print( pddf[ : 7 ] )
print()
print( pddf['Series1'] )
print()
print( f"len( pddf['Series1'] ) = {len( pddf['Series1'] )} " )
print()
print(f"len( pddf['Series1'] )//2 = {len( pddf['Series1'] )//2}")
print()
print( pddf['Series1'][ : len( pddf['Series1'] )//2 ] )
```

```python
#We can store the sliced data as another column in DataFrame using direct method.
s1 = pddf['Series1']
print( s1 )
print()
pddf['halfofSeries1'] = s1[ : len( pddf['Series1'] ) // 2 ]
pddf
```

```python
#we can see that, half Series column has only half data of Series1 and after half data, everything is null.
```

```python
temp = pddf.copy()

temp.pop( 'halfofSeries1' )

print( temp )
```

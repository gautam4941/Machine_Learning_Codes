```python
import pandas as pd
```

```python
data = pd.read_csv( 'Excel Data/Social_Network_Ads.csv' )
data.head()
```

```python
"""
Syntax :-  pd.get_dummies( column_name )
"""
pd.get_dummies( data['Gender'] )
```

```python
#Storing get_dummies value into a variable and then, we will concatinate the dummy_data with our dataset
```

```python
gender_dummy = pd.get_dummies( data['Gender'] )
gender_dummy
```

```python
#Concation of one dataFrame with another -> pd.concat( )

#Syntax :- pd.concat( [ DataFrame1, DataFrame2], axis = columns )

pd.concat( [ data, gender_dummy ], axis = 'columns'  )

#or

#pd.concat( [ data, gender_dummy ], axis = 1  )
```

```python
pd.concat( [ data, gender_dummy ], axis = 'rows'  )

#or

#pd.concat( [ data, gender_dummy ], axis = 0  )
```

```python
data = pd.concat( [ data, gender_dummy ], axis = 'columns'  )

data
```

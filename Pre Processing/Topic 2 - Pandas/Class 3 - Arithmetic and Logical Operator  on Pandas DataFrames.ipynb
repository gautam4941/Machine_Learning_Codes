```python
#Arithmetic Addition

for i in range( len( pddf['Series1'].index ) ):    #0,1, 2, ..., 28
    
    print( f"pddf['Series1'][{i}] = { pddf['Series1'][i] }", end = ', ' )
    print( f"pddf['Series2'][{i}] = { pddf['Series2'][i] }", end = ', ' )
    print( f"pddf['Series1'][{i}] + pddf['Series2'][{i}] = { pddf['Series1'][i] + pddf['Series2'][i] }" )
    print()
```

```python
#Arithmetic Multiplication

for i in range( len( pddf['Series1'] ) ):
    
    print( f"pddf['Series1'][{i}] = { pddf['Series1'][i] }", end = ', ' )
    print( f"pddf['Series2'][{i}] = { pddf['Series2'][i] }", end = ', ' )
    print( f"pddf['Series1'][{i}] * pddf['Series2'][{i}] = { pddf['Series1'][i] * pddf['Series2'][i] }" )
    print()
```

```python
#Arithmetic Division

for i in range( len( pddf['Series1'] ) ):
    
    print( f"pddf['Series1'][{i}] = { pddf['Series1'][i] }", end = ', ' )
    print( f"pddf['Series2'][{i}] = { pddf['Series2'][i] }", end = ', ' )
    print( f"pddf['Series1'][{i}] / pddf['Series2'][{i}] = { pddf['Series1'][i] / pddf['Series2'][i] }" )
    print()
```

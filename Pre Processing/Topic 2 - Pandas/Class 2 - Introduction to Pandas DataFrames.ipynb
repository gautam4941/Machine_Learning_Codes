```python
import pandas as pd
```

```python
pds1 = pd.Series( data = list( range(1, 10, 2) ), index = ['one', 'three', 'five', 'seven', 'nine'] )

pds2 = pd.Series( data = list( range(10, 20, 2) ), index = [ 'ten', 'twelve', 'forteen', 'sixteen', 'eighteen' ] )

pds3 = pd.Series( data = list( range(20, 30, 2) ), index = ['twenty', 'twentytwo', 'twentyfour', 'twentysix', 'twentyeight' ] )

pds4 = pd.Series( data = list( range(1, 10, 1) ), index = ['one', 'two', 'three', 'four', 'five', 'six', 'seven', 'eight', 'nine'] )

pds5 = pd.Series( data = list( range(10, 20, 1) ), index = [ 'ten', 'eleven', 'twelve', 'thirteen', 'forteen', 'fifteen', 'sixteen', 'seventeen', 'eighteen', 'nineteen' ] )

pds6 = pd.Series( data = list( range(20, 30, 1) ), index = ['twenty', 'twentyone', 'twentytwo', 'twentythree', 'twentyfour', 'twentyfive', 'twentysix', 'twentyseven', 'twentyeight',  'twentynine' ] )

print( "pds1, " )
print( pds1 )
print()
print( "pds2, " )
print( pds2 )
print()
print( "pds3, " )
print( pds3 )
print()
print( "pds4, " )
print( pds4 )
print()
print( "pds5, " )
print( pds5 )
print()
print( "pds6, " )
print( pds6 )
```

```python
#Now we will merge all the Pandas Series into a group using dictionary. Where each Panda Series will be having a label.
# and this labelling thing can only be done with the help of Dictionary.
```

```python
d = { 'Series1' : pds1, 'Series2' : pds2, 'Series3' : pds3, 'Series4' : pds4, 'Series5' : pds5, 'Series6' : pds6  }

print( d )
print()
print( f"type(d) = { type(d) }" )
print()

print( "d['Series2']" )
print( d['Series2'] )
print()

for i in d:
    print( f"i = {i}" )  #Series1, Series2, ...
    print( d[i] )        #d[ Series1 ], d[ Series2 ], ...
    print()
```

```python
#Now still our data in dictionary can be said as data Frame as, it is not arrange in the form of row and col.
#To do that, we will have to convert dicitionary( d ) into Data Frame using type Casting.
print(d)
```

```python
pddf = pd.DataFrame( d )   #Typecasting dicitonary into DataFrame
print(pddf)

#Now we can see the diffrence between dictionary and DataFrame.
#NaN is null value here.
```

```python
#To see the indexes of DataFrame.
row = pddf.index
print( row )
print()
col = pddf.columns
#To see the columns of DataFrame.
print( col )
print()
print( f"type(row) = { type(row) }" )       #<class 'pandas.core.indexes.base.Index'>    ----> DataType is itself Index
print( f"type(col) = { type(col) }" )       #<class 'pandas.core.indexes.base.Index'>    ----> DataType is itself Index
```

```python
#Columns of DataFrame,

pddf.columns = [1, 2, 3, 4, 5, 6]
'''
It will Update the column names of Pandas DataFrame if columns is already created before  --> Partially Wrong Statement.
There is no if for columns because, In dataFrame columns are the keys of dictionary and dictionary can't be created without keys
'''
print( pddf )
```

```python
#Indexes of DataFrame,
old_index = pddf.index
print( old_index )
print()
print(f"list( range( 1, 30, 1 ) ) = {list( range( 1, 30, 1 ) )}" )
print()

pddf.index = list( range( 1, 30, 1 ) )   #Explicit Method to assign Index
'''
It will Update the indexes( rows ) of Pandas DataFrame if indexes( rows ) are already created before.
Otherwise, we can give index names in this way also.
'''
print( pddf )
```

```python
#Updating index of DataFrame to oldindex

pddf.index = old_index
print( pddf )
```

```python
'''
1) Setting Index using set_index()
2) head()
3) tail()
4) Matplotlib
'''

import pandas as pd
```

```python
l1 = ['Mon', 'Tue', 'Wed', 'Thr', 'Fri', 'Sar', 'Sun']
l2 = [8, 6, 9, 4, 1, 0]
l3 = [2, 1, 1, 2, 0, 3]
l4 = [2, 2, 4, 3, 0, 2]
l5 = [6, 8, 6, 7, 4, 9]

print(f"l1 = {l1}")
print(f"l2 = {l2}")
print(f"l3 = {l3}")
print(f"l4 = {l4}")
print(f"l5 = {l5}")

pds1 = pd.Series( l1 )
pds2 = pd.Series( l2 )
pds3 = pd.Series( l3 )
pds4 = pd.Series( l4 )
pds5 = pd.Series( l5 )

print()

print( pds1 )
print( pds2 )
print( pds3 )
print( pds4 )
print( pds5 )
```

```python
d = {'Days' : pds1, 'Studying' : pds2, 'Television' : pds3, 'Playing' : pds4, 'Sleeping' : pds5}

print( d )
print()
print("Printing d['Days'], ")
print( d['Days'] )                 #dict_name[ key ] -> VALUE ASSOCIATED TO THAT KEY

print()
for i in d:                            #-> Keys :- Days
    print(i, end = ' -> \n')
    print( d[i] )                      #-> d['Days']
```

```python
pddf = pd.DataFrame( d )

print(f" pddf  = \n{pddf } \nand type( pddf  ) = { type( pddf  ) } ")
```

```python
temp = pddf.copy()

print( temp['Days'] )
print()
print( "DataFrame Indexes, " )
print( f"temp.index = { temp.index }" )
print()
print( f"list( temp.index ) = { list( temp.index ) }" )
print( f"tuple( temp.index ) = { tuple( temp.index ) }" )
print( f"set( temp.index ) = { set( temp.index ) }" )
print()

temp.index = temp['Days']
print( temp )
print()
temp.pop('Days')
print( temp )
```

```python
#We have a problem that there is no name for index. We need to make string as index.

#Let's make Days( col1 ) and it's value as index. To do it we need a function named as set_index( ).
#Syntax :-

#dataFrame_name.set_index( column_name )

print( temp )
temp = pddf.copy()
print()
print( temp )
print()
temp.set_index( 'Days' )
print( temp )
print()
temp.set_index( 'Days', inplace = False )
print( temp )
print()
temp.set_index( 'Days', inplace = True )
print( temp )
#DataFrame_name.set_index( 'col_name', inplace )

#pddf.set_index('Days', inplace = False) # our data frame will not be updated if inplace is False.
#print( pddf )
```

```python
print( pddf )
print()
pddf.set_index('Days', inplace = True ) # our data frame will be updated if inplace is True.
print( pddf )
```

```python
#head and tail,

#why we need head() and tail() - we can analyze small data by printing but if the data is very big.
#Printing is not a very good idea, as it takes a lot of space + time to execute print statement.

#To overcome this issue, we got head() and tail(). head() will show the top 5 rows of our dataFrame whereas, tail() will 
#show the last 5 data of our dataFrame.
```

```python
#head - head() are of 2 types. a) Parametric head() and b) default head()

#Parametric head() - this will allow us to see data till n rows and 1<=n<=rows_of_dataFrame.
#syntax :-
# DataFrame_name.head( no_of_rows )

print( pddf.head( 3 ) )
print()

#Non Parametric head() - this will allow us to see data till 5 rows
print( pddf.head() )
print()
print( pddf.head( 6 ) )
print()
print( pddf.head( 100 ) )
```

```python
#tail - tail() are of 2 types. a) Parametric tail() and b) default tail()

#Parametric tail() - this will allow us to see data till n rows and rows_of_dataFrame<=n<=1 .
#syntax :-
# DataFrame_name.tail( no_of_rows )

print( pddf.tail( 3 ) )
print()

#Non Parametric head() - this will allow us to see data till 5 rows
print( pddf.tail() )
print()
print( pddf.tail( 6 ) )
```

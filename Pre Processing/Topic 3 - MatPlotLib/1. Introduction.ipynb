```python
import matplotlib.pyplot as plt             #Here, plt is the alias name of matplotlib.pyplot
```

```python
x = [1,2,3,4,5,6,7]
y = [50,51,52,48,47,49,46]
z = [ 12000, 15000, 10000, 2000, 7000, 18000, 20000]

print(x)
print()
print(y)
print()
print(z)
```

```python
plt.plot( x, y )
```

```python
plt.plot( x, y, color = 'red' )
```

```python
plt.plot( x, y, color = 'red', marker = '*' )
```

```python
plt.plot( x, y, color = 'red', marker = '*', linestyle = '' )
```

```python
plt.plot( x, y, color = 'blue', marker = '*' )
plt.xlabel('X label')
plt.ylabel('Y label')
```

```python
#plot()
#Syntax :-
#plt.plot( dataset1, dataset2, marker = 'character', color = colorname )

#After dataset1 and dataset2. All parameters are optional

#plot() works with list, tuples, Series and DataFrames.


print(f"x = {x}")
print(f"y = { y }")
print(f"z = {z}")
print()
plt.plot( x, y, marker = '*', color = 'red' )
plt.plot( x, z, marker = 'o', color = 'blue' )
#label()

#1) xlabel() 
#Syntax :-  plt.xlabel( label_name )

#2) ylabel
#Syntax :- plt.ylabel( label_name )

plt.xlabel( 'S. NO' )
plt.ylabel( 'Salary' )
plt.title('Graph fror Salary vs S. No')
plt.show()
```

```python
# Same effect as 'b+' format string
plt.plot(x, y, color='blue',marker='+',linestyle = '')
```

```python
plt.plot(x,y,color='blue',marker='*',linestyle='', markersize = 20 )
```

```python
import numpy as np
```

```python
x = np.linspace( 0, 10, 100 )
x
```

```python
plt.plot(x, np.sin(x), label = 'Sin')
plt.plot(x, np.cos(x), label = 'Cos')
plt.legend()
plt.show()
```

```python
#create the first of two panels and set current axis

plt.subplot( 2, 1, 1 ) # (rows, columns, panel number)
plt.plot( x, np.sin(x), label = 'Sin' )
plt.legend()

# create the second panel and set current axis
plt.subplot( 2, 1, 2 )
plt.plot( x, np.cos(x), label = 'Cos' )

plt.legend()
plt.show()
```

```python
#create the first of two panels and set current axis

plt.subplot( 2, 2, 1 ) # (rows, columns, panel number)
plt.plot( x, np.sin(x), label = 'Sin' )
plt.legend()

# create the second panel and set current axis
plt.subplot( 2, 2, 4 )
plt.plot( x, np.cos(x), label = 'Cos' )

plt.legend()
plt.show()
```

```python
ax = plt.axes()
x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x))
plt.show()
```

```python
plt.plot(x, x + 0, label = 'x + 0', linestyle='solid')
plt.plot(x, x + 1, label = 'x + 1', linestyle='dashed')
plt.plot(x, x + 2, label = 'x + 2', linestyle='dashdot')
plt.plot(x, x + 3, label = 'x + #', linestyle='dotted');
# For short, you can use the following codes:
plt.plot(x, x + 4, label = 'x + 4', linestyle='-') # solid
plt.plot(x, x + 5, label = 'x + 5', linestyle='--') # dashed
plt.plot(x, x + 6, label = 'x + 6', linestyle='-.') # dashdot
plt.plot(x, x + 7, label = 'x + 7', linestyle=':'); # dotted

plt.legend( loc = 'best' )
plt.show()
```

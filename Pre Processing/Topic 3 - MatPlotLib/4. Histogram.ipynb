```python
import matplotlib.pyplot as plt
import random
```

```python
#Syntax for Random,

#random.randrange(start_value, End_value, incrementation)

#random.randrange( 20, 50, 3 ) -> 20, 23, 26, 29, ... , 48

roll_no = []

for i in range(1, 41 ):                                 #1, 2, .., 39, 40
    roll_no.append( random.randrange( 20, 50, 1 ) )
    
print( f"roll_no = { roll_no }" )

count = list( range( 1, 41 ) )               #1, 2, ..., 40

print( f"\ncount = { count }" )

print( f"\nlen(  roll_no ) = { len(  roll_no ) }" )
print( f"len(  count ) = { len( count ) }" )
```

```python
plt.bar( roll_no, count , label = 'Roll_no V/s Count')
plt.xlabel('roll_no')
plt.ylabel('count')
plt.title( 'Histogram Graph' )
plt.legend()
plt.show()
```

```python
plt.hist( roll_no, count , label = 'Roll_no V/s Count')
plt.xlabel('roll_no')
plt.ylabel('count')
plt.title( 'Histogram Graph' )
plt.legend()
plt.show()
```

```python
#https://keydifferences.com/difference-between-histogram-and-bar-graph.html
```

```python
plt.hist( roll_no, count , label = 'Roll_no V/s Count', rwidth = 0.9 )  #rwidth can't be more than 1
plt.xlabel('roll_no')
plt.ylabel('count')
plt.title( 'Histogram Graph' )
plt.legend()
plt.show()
```

```python
plt.hist( roll_no, count , label = 'Roll_no V/s Count', rwidth = 0.5 )

plt.xlabel('roll_no')
plt.ylabel('count')
plt.title( 'Histogram Graph' )
plt.legend()
plt.show()
```

```python
plt.hist( roll_no, count , label = 'Roll_no V/s Count', rwidth = 20 )
plt.xlabel('roll_no')
plt.ylabel('count')
plt.title( 'Histogram Graph' )
plt.legend()
plt.show()
```

```python
import pandas as pd
```

```python
data = pd.read_csv( r"hrdata.csv" )
print( f"data.shape = { data.shape }" )
data.head()
```

```python
# x : label or position, default None
# Only used if data is a DataFrame.

data.plot( x = 'Income' )
```

```python
# y : label, position or list of label, positions, default None
# Allows plotting of one column versus another. Only used if data is a DataFrame.

data.plot( y = 'Income' )
```

```python
data.plot( x = 'Income', y = 'Age' )
```

```python
# kind : str
# The kind of plot to produce: 
# â€˜lineâ€™ : line plot (default)

data.plot( x = 'Income', y = 'Age', kind = 'line' )
```

```python
# â€˜barâ€™ : vertical bar plot

data.plot( x = 'Income', y = 'Age', kind = 'bar' )
```

```python
# â€˜barhâ€™ : horizontal bar plot

data.plot( x = 'Income', y = 'Age', kind = 'barh' )
```

```python
#â€˜histâ€™ : histogram

data.plot( x = 'Income', y = 'Age', kind = 'hist' )
```

```python
# â€˜boxâ€™ : boxplot

data.plot( x = 'Income', kind = 'box' )
```

```python
# â€˜boxâ€™ : boxplot

data.plot( y = 'Age', kind = 'box' )
```

```python
# â€˜boxâ€™ : boxplot

data.plot( x = 'Income', y = 'Age', kind = 'box' )
```

```python
# â€˜kdeâ€™ : Kernel Density Estimation plot

data.plot( x = 'Income', y = 'Age', kind = 'kde' )
```

```python
# â€˜densityâ€™ : same as â€˜kdeâ€™

data.plot( x = 'Income', y = 'Age', kind = 'density' )
```

```python
# â€˜areaâ€™ : area plot

data.plot( x = 'Income', y = 'Age', kind = 'area' )
```

```python
# â€˜pieâ€™ : pie plot

data.plot( x = 'Income', y = 'Age', kind = 'pie' )
```

```python
# â€˜scatterâ€™ : scatter plot

data.plot( x = 'Income', y = 'Age', kind = 'scatter' )
```

```python
# â€˜hexbinâ€™ : hexbin plot.

data.plot( x = 'Income', y = 'Age', kind = 'hexbin' )
```

```python
# figsize : a tuple (width, height) in inches
# use_index : bool, default True
# Use index as ticks for x axis.

data.plot( x = 'Income', y = 'Age', kind = 'scatter', figsize = (10, 10) )
```

```python
# title : str or list
# Title to use for the plot. If a string is passed, print the string at the top of the figure. 
# If a list is passed and subplots is True, print each item in the list above the corresponding subplot.

data.plot( x = 'Income'
          , y = 'Age'
          , kind = 'scatter'
          , figsize = (10, 10)
          , title = 'Pandas_Graphs' )
```

```python
# grid : bool, default None (matlab style default)
# Axis grid lines.

data.plot( x = 'Income'
          , y = 'Age'
          , kind = 'scatter'
          , figsize = (10, 10)
          , title = 'Pandas_Graphs'
          , grid = None )
```

```python
data.plot( x = 'Income'
          , y = 'Age'
          , kind = 'scatter'
          , figsize = (10, 10)
          , title = 'Pandas_Graphs'
          , grid = False )
```

```python
data.plot( x = 'Income'
          , y = 'Age'
          , kind = 'scatter'
          , figsize = (10, 10)
          , title = 'Pandas_Graphs'
          , grid = True )
```

```python
# logx : bool or â€˜symâ€™, default False
# Use log scaling or symlog scaling on x axis. .. versionchanged:: 0.25.0

data.plot( x = 'Income'
          , y = 'Age'
          , kind = 'scatter'
          , figsize = (10, 10)
          , title = 'Pandas_Graphs'
          , grid = True
          , logx = 'sym' )
```

```python
# logy : bool or â€˜symâ€™ default False
# Use log scaling or symlog scaling on y axis. .. versionchanged:: 0.25.0

data.plot( x = 'Income'
          , y = 'Age'
          , kind = 'scatter'
          , figsize = (10, 10)
          , title = 'Pandas_Graphs'
          , grid = True
          , logx = False )
```

```python
# loglog : bool or â€˜symâ€™, default False
# Use log scaling or symlog scaling on both x and y axes. .. versionchanged:: 0.25.0

data.plot( x = 'Income'
          , y = 'Age'
          , kind = 'scatter'
          , figsize = (10, 10)
          , title = 'Pandas_Graphs'
          , grid = True
          , logx = True )
```

```python
# fontsize : int, default None
# Font size for xticks and yticks.

data.plot( x = 'Income'
          , y = 'Age'
          , kind = 'scatter'
          , figsize = (10, 10)
          , title = 'Pandas_Graphs'
          , grid = True
          , fontsize = 20 )
```

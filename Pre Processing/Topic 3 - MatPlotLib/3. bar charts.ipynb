```python
company = ['Google', 'Yahoo', 'Accenture', 'Philips']
revenue = [100, 50, 50, 10]

print( f"company = {company}" )
print( f"revenue = { revenue }" )
```

```python
import numpy as np
```

```python
print( f"len( company ) = {len( company )}" )

xpos = np.arange( len( company ) )                                 #np.arange( 4 )
print( f"type(xpos) = {type(xpos)} and xpos = {xpos}" )
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(company, revenue, label="Revenue")  #We are using bar instead of plot.

plt.xlabel("Companies")
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
plt.show()
```

```python
plt.bar(xpos, revenue, label="Revenue")    #We are using bar instead of bar.

plt.xticks(xpos,company)                #we use xticks to replace the x axis data. Syntax:- plt.xticks( old_X, new_X )
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
plt.show()
```

```python
profit=[40,2,34,12]

plt.bar(xpos,revenue, label="Revenue")
plt.bar(xpos,profit, label="Profit")

plt.xticks(xpos,company)
plt.xlabel("xpos")
plt.ylabel("Revenue(Bln) and Proit")
plt.title('US Technology Stocks')
plt.legend()
```

```python
profit=[40,2,34,12]

plt.bar(xpos,revenue, width=0.4, label="Revenue")
plt.bar(xpos,profit, width=0.6,label="Profit")

plt.xticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
```

```python
#Syntax for plt.barh( y_axis, x_axis )

plt.barh(xpos,revenue, label="Revenue")   # ".bar()" -> is for Displaying the bar vertically( alignment w.r.t to y axis ),
                                          # Whereas, ".barh()" ->  To display bars Horizontally( alignment w.r.t to x axis)

plt.yticks(xpos,company)
plt.xlabel( 'Stocks' )
plt.ylabel("Revenue(Bln)")
plt.title('US Technolog Stocks')
plt.legend()
```

```python
'''
ML TIP :-

So, It's good to follow, 

i) .bar() +  .yticks() -> Improper
ii) .barh() +  .xticks() -> Improper

iii) .bar() +  .xticks() -> proper
iv) .barh() +  .yticks() -> proper
'''
```

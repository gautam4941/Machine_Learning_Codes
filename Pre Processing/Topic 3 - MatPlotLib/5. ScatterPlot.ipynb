```python
import matplotlib.pyplot as plt
import random
```

```python
roll_no = []

for i in range( 40 ):
    roll_no.append( random.randrange( 20, 50, 1 ) )
    
print( f"roll_no = { roll_no }" )

count = list( range( 1, 41 ) )

print( f"\ncount = { count }" )

print( f"\nlen(  roll_no ) = { len(  roll_no ) }" )
print( f"len(  count ) = { len( count ) }" )
```

```python
plt.scatter( roll_no, count )
```

```python
plt.scatter( roll_no, count , label = 'Roll_no V/s Count')
```

```python
plt.scatter( roll_no, count , label = 'Roll_no V/s Count')
plt.xlabel( 'Roll_no' )
plt.ylabel( 'Count' )
```

```python
plt.scatter( roll_no, count , label = 'Roll_no V/s Count')
plt.xlabel( 'Roll_no' )
plt.ylabel( 'Count' )
plt.title( 'Scatter Plot' )
```

```python
plt.scatter( roll_no, count , label = 'Roll_no V/s Count')
plt.xlabel( 'Roll_no' )
plt.ylabel( 'Count' )
plt.title( 'Scatter Plot' )
plt.legend()
```

```python
plt.scatter( roll_no, count , label = 'Roll_no V/s Count')
plt.xlabel( 'Roll_no' )
plt.ylabel( 'Count' )
plt.title( 'Scatter Plot' )
plt.legend()
plt.show()
```

```python
plt.scatter( roll_no, count , label = 'Roll_no V/s Count', color = 'red')
plt.xlabel( 'Roll_no' )
plt.ylabel( 'Count' )
plt.title( 'Scatter Plot' )
plt.legend()
plt.show()
```

```python
plt.scatter( roll_no, count , label = 'Roll_no V/s Count', color = 'red', marker = '*')
plt.xlabel( 'Roll_no' )
plt.ylabel( 'Count' )
plt.title( 'Scatter Plot' )
plt.legend()
plt.show()
```

```python
plt.scatter( roll_no, count , label = 'Roll_no V/s Count', color = 'red', marker = '*', s = 30) #Here, s is Size of marker.
plt.xlabel( 'Roll_no' )
plt.ylabel( 'Count' )
plt.title( 'Scatter Plot' )
plt.legend()
plt.show()
```

```python
plt.plot( roll_no, count , label = 'Roll_no V/s Count', color = 'red', marker = '*', linestyle = '') #Here, s is Size of marker.
plt.xlabel( 'Roll_no' )
plt.ylabel( 'Count' )
plt.title( 'Scatter Plot' )
plt.legend()
plt.show()
```

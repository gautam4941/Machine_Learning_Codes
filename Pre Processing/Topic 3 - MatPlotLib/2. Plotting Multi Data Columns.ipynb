```python
import matplotlib.pyplot as plt
```

```python
days =  [1,2,3,4,5,6,7]
max_t = [50,51,52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]

print( f"days = { days }" )
print( f"max_t = { max_t }" )
print( f"min_t = { min_t }" )
print( f"avg_t = { avg_t }" )
```

```python
plt.plot(days, max_t, color = 'red' )
plt.xlabel('days')
plt.ylabel('Max_t')
plt.title('Days v/s Mat Temperature')
plt.show()
```

```python
plt.plot(days, min_t, color = 'red' )
plt.xlabel('days')
plt.ylabel('Max_t')
plt.title('Days v/s Min_t')
plt.show()
```

```python
plt.plot(days, avg_t, color = 'red' )
plt.xlabel('days')
plt.ylabel('Max_t')
plt.title('Days v/s Avg_t')
plt.show()
```

```python
plt.plot(days, max_t, color = 'green')
plt.plot(days, min_t, color = 'red')
plt.plot(days, avg_t, color = 'blue')
plt.xlabel('Days')
plt.ylabel('Max_t, Min_t and Avg_t')
plt.title('Temperature V/s Days')
plt.show()
```

```python
plt.plot(days, max_t, label="max", color = 'green')
plt.plot(days, min_t, label="min", color = 'red')
plt.plot(days, avg_t, label="average", color = 'blue')
plt.xlabel('Days')
plt.ylabel('Max_t, Min_t and Avg_t')
plt.title('Temperature V/s Days')
plt.legend()
plt.show()
```

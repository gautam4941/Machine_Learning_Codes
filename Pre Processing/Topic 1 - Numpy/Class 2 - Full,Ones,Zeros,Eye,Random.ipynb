```python
'''Things going to be covered, 
1) full   -> To fill Custom value of any data type.
2) ones   -> To fill 1. value
3) zeros   -> To fill 0. value
4) eye   -> Identity Matrix
5) random.random  -> To fill random values/ numbers from 0 to 1
'''
```

```python
import numpy as np
```

```python
#fill - To fill the array with same elements( Ex :- fill the array elements as 7 everywhere in array )
#full() :- np.full( dim, value )

arr7 = np.full( (3, 2, 3), 8 )
print( arr7 )
```

```python
#fill - To fill the array with same elements( Ex :- fill the array elements as 7.6 everywhere in array )
arr7point6 = np.full( (2, 4), 7.6 )
print( arr7point6 )
print()
arrone = np.full( (2, 4), 'Hi' )
print(arrone)
```

```python
#zeros - To fill the array with 0( Ex :- fill the array elements as 0 everywhere in array )
#zeros :- np.zeros( dim )

arr0 = np.zeros( (2, 4) )
print( arr0 )
print()
arr0 = np.full( (2, 4), 0. )
print( arr0 )
```

```python
#ones - To fill the array with 1( Ex :- fill the array elements as 1 everywhere in array )
#zeros :- np.ones( dim )

arr1 = np.ones( (2, 4) )
print( arr1 )
print()
arr1 = np.full( (2, 4), 1. )
print( arr1 )
```

```python
#eye - To fill the array with digonal as 1( Ex :- fill the array elements digonally as 1 and others as 0 in array )

#Syntax :- np.eye( dim )

arreye = np.eye( 5 )    #Only Single '()' bracket. No double bracket like fill, zeros and ones
print( arreye)
```

```python
#random.random - To fill the array with random number form 0 to 1
#Syntax :- np.random.random( dim )

arrrandom = np.random.random( (2, 4) )
print( f"arrrandom :- \n{ arrrandom }\n" )

#Syntax :- np.random.randint( range1, range2, dim )

n1 = np.random.randint( 2, 10, 4 )
n2 = np.random.randint( 12, 20, 4 )
n3 = np.random.randint( 22, 30, (4, 3) )

print( f"n1 :- \n{ n1 }\n" )
print( f"n2 :- \n{ n2 }\n" )
print( f"n3 :- \n{ n3 }\n" )
```

```python
#Syntax :- np.random.seed( seed_value )
print( f"1. Without Seed, np.random.randint( 2, 10, 4 ) = { np.random.randint( 2, 10, 4 ) }")
print( f"2. Without Seed, np.random.randint( 2, 10, 4 ) = { np.random.randint( 2, 10, 4 ) }")
print( f"3. Without Seed, np.random.randint( 2, 10, 4 ) = { np.random.randint( 2, 10, 4 ) }")
```

```python
np.random.seed( 234234 )
print( f"4. With Seed, np.random.randint( 2, 10, 4 ) = { np.random.randint( 2, 10, (4, 100, 2, 5) ) }")
```

```python
print( f"5. Without Seed, np.random.randint( 2, 10, 4 ) = { np.random.randint( 2, 10, 4 ) }")

print( f"6. With Seed, np.random.randint( 2, 10, 4 ) = { np.random.randint( 2, 10, 4 ) }")

print( f"7. With Seed, np.random.randint( 2, 10, 4 ) = { np.random.randint( 2, 10, 4 ) }")
```

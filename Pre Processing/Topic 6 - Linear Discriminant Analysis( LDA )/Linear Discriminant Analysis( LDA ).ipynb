```python
import pandas as pd
```

```python
data =  pd.read_csv('winequality_red.csv')
data.head()
```

```python
data.isnull().sum()
```

```python
x = data.loc[ :, 'fixed acidity' : 'alcohol' ]
y = data.loc[ :, 'quality' ]
```

```python
x.head()
```

```python
y.head()
```

```python
data.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split( x, y, test_size = 0.2 )
```

```python
x_train.head()
```

```python
x_test.head()
```

```python
y_train.head()
```

```python
y_train.head()
```

```python
from sklearn.preprocessing import StandardScaler

ssc = StandardScaler()
```

```python
x_train = ssc.fit_transform( x_train )
x_test = ssc.fit_transform( x_test )
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
```

```python
lda = LDA( n_components = 3 )
```

```python
lda.fit_transform( x_train, y_train )
```

```python
lda = LDA( n_components = 2 )
```

```python
x_train = lda.fit_transform( x_train, y_train )
x_train
```

```python
x_test = lda.transform( x_test )
x_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit( x_train, y_train )
```

```python
y_pred = lr.predict( x_test )
y_pred
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot( x_test, y_test, color = 'blue', marker = 'o', label = 'y_test', linestyle = '' )
plt.plot( x_test, y_pred, color = 'red', marker = '*', label = 'y_pred', linestyle = '' )
plt.xlabel( 'X-test' )
plt.ylabel( 'Y_test V/s Y-pred' )
plt.title( 'LR After LDR, ' )
plt.legend()
plt.show()
```

```python
lr.score( x_test, y_test )
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix( y_test, y_pred )
```

```python
cm
```

```python
print( y_pred )
print()
print( y_pred.shape )
print()
row = y_pred.shape[0]
print( f"row = { row }" )
```

```python
accuracy_sum = 0
error_sum = 0

for i in range( len(cm ) ):
    for j in range( len( cm[i] ) ):
        if( i == j ):
            accuracy_sum = cm[i][j] + accuracy_sum
        else:
            error_sum = error_sum + cm[i][j]
```

```python
print( f"accuracy_sum = { accuracy_sum }" )
print()
print( f" Accuracy % = { accuracy_sum/row }" )
```

```python
print( f"error_sum = { error_sum }" )
print()
print( f" Error % = { error_sum/row }" )
```

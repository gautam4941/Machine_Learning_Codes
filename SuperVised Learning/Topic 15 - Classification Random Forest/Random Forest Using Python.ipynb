```python
import pandas as pd
```

```python
data = pd.read_csv( "Social_Network_Ads.csv" )
```

```python
data.head()
```

```python
data['Purchased'].unique()    # Purchased is 0 and 1 . So, it is Classification data.
```

```python
data.isnull().sum()
```

```python
data = data.drop( ['User ID'], axis = 1 )
```

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
data['Gender'] = le.fit_transform( data['Gender'] )
```

```python
data.head()
```

```python
data.shape
```

```python
x = data.iloc[ : , : 3] #0, 1, 2
y = data.iloc[ : , 3]

x.head()
```

```python
y.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ssc = StandardScaler()
```

```python
x = ssc.fit_transform( x )
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split( x, y, test_size = 0.2 )  #train_data = 0.8 = 80% = 320 rows
```

```python
print( len( x_train ) )

x_train
```

```python
print( len( x_test ) )

x_test
```

```python
print( len( y_train ) )

y_train
```

```python
print( len( y_test ) )

y_test
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
?RandomForestClassifier
```

```python
rfc = RandomForestClassifier( criterion = 'entropy' )
```

```python
rfc.fit( x_train, y_train )
```

```python
y_pred = rfc.predict( x_test )
```

```python
y_pred
```

```python
rfc.score( x_test, y_test )             #Here, Prediction accuract will change as randomly data get allocated to train and test data.
```

```python
from sklearn.metrics import confusion_matrix    #Hence, it is classification Dataset . We are checking accuracy using Confusion Matrix
```

```python
print( f"len( x_test ) = { len( x_test ) }" )
print( f"len( x_test ) = { len( x_test ) }" )
```

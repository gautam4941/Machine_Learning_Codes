```python
import pandas as pd
```

```python
data = pd.read_csv("Social_Network_Ads.csv")
data
```

```python
data.isnull().sum()
```

```python
data.shape
```

```python
data = data.drop( ['User ID'], axis = 1 )
```

```python
data.head()
```

```python
data.shape
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
data['Gender'] = le.fit_transform( data['Gender'] )
```

```python
data.head()
```

```python
x = data.iloc[ :, : 3 ]
y = data.iloc[ :, 3]
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split( x, y, test_size = 0.2 )
```

```python
print( x_train.shape )

x_train.head()
```

```python
print( x_test.shape )

x_test.head()
```

```python
print( y_train.shape )

y_train.head()
```

```python
print( y_test.shape )

y_test.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ssc = StandardScaler()
```

```python
x_train = ssc.fit_transform( x_train )
x_test = ssc.fit_transform( x_test )
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier( criterion = 'entropy' )
dtc
```

```python
dtc.fit( x_train, y_train )
```

```python
data
```

```python
print( f"dtc.predict( [[0, 13, 2000]] ) = { dtc.predict( [[0, 13, 2000]] ) }" )
```

```python
y_pred = dtc.predict( x_test )
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix( y_test, y_pred )
```

```python
cm
```

```python
accuracy = cm[0][0] + cm[1][1]
loss = cm[0][1] + cm[1][0]

print( f"accuracy = { accuracy }, loss = { loss } and accuracy % = { accuracy/( accuracy + loss ) }" )
```

```python
dtc.score( x_test, y_test )             #Here, Prediction accuract will change as randomly data get allocated to train and test data.
```

```python
import matplotlib.pyplot as plt
```

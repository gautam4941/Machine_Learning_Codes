```python
y = data['Creditability']                    #y has only 'Creditability' column data.
x = data.loc[ :, 'Account Balance' :  ]    #x has 20 columns of data.
```

```python
x.head()
```

```python
y.head()
```

```python
print(f"len(x) = {len(x)}")
print(f"len(y) = {len(y)}")
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1 )   #train_daata = 0.9 = 90% = 900 rows.
```

```python
print( f"x_train.head() = {x_train.head()}" )
print( f"\n\nx_test.head() = {x_test.head()}" )
print( f"\n\ny_train.head() = {y_train.head()}" )
print( f"\n\ny_test.head() = {y_test.head()}" )
```

```python
print(f"len( x_train ) = { len( x_train ) }")
print(f"len( x_test ) = { len( x_test ) }")
print(f"len( y_train ) = { len( y_train ) }")
print(f"len( y_test ) = { len( y_test) }")
```

```python
from sklearn.naive_bayes import GaussianNB   #Importing Library for Naive Bayes.
```

```python
gnb = GaussianNB()                           #Making Objext of GaussianNB class.

gnb.fit(x_train,y_train)                    #Giving training using x_train and y_train

y_pred = gnb.predict(x_test)                    #We are finding the new value. using predict().

#y_test is not used.
#y_test v/s new_y_test.
```

```python
print( f"len( y_test ) = { len( y_test ) }" )

y_test.head()
```

```python
print( f"y_pred = { y_pred }" )
print( f"len( y_pred ) = { len( y_pred ) }" )
```

```python
print( y_test.unique() )
print()
print( pd.Series( y_pred ).unique() )
```

```python
gnb.score(x_test, y_test)    #To Find the Accuracy of our predicted data.  It Always accept 2d array not 1d -
    
#dimensional data.
```

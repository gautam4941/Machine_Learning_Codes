```python
import pandas as pd
```

```python
data = pd.read_csv('../DataSet/Social_Network_Ads.csv')
data.head()
```

```python
data.isnull().sum()
```

```python
data.shape
```

```python
gender_dummy = pd.get_dummies( data['Gender'] )
print( gender_dummy.head() )
```

```python
data = pd.concat( [ data, gender_dummy ], axis = 'columns' )
print( data.head() )
```

```python
final_data = data.drop( ['Gender'], axis = 'columns' )
final_data.head()
```

```python
y = final_data[ 'EstimatedSalary' ]
x = final_data.drop( 'EstimatedSalary', axis = 'columns' )
```

```python
print(x.head())
print()
print(y.head())
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
print( f"regression = {regression}" )
```

```python
regression.fit( x, y )
```

```python
y_pred = regression.predict( x )
y_pred
```

```python
regression.score( x , y )
```

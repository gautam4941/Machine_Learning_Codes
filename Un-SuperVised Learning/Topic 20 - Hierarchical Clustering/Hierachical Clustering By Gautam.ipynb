```python
import pandas as pd
```

```python
data = pd.read_csv('Mall_Customers.csv')
data.head()
```

```python
data.isnull().sum()
```

```python
x = data.loc[ :, 'Annual Income (k$)' : 'Spending Score (1-100)' ]
x.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot( x, linestyle = '', marker = '*' )
plt.title('Annual Income (k$) and Spending Score (1-100)')
plt.show()

#Here, Blue Line represents x['Annual Income (k$)'] where as yellow points represents, x['Spending Score (1-100)']
#Analyze the Dataset.
```

```python
from scipy.cluster import hierarchy as sch
```

```python
print( f"len( sch.linkage( x ) ) = { len( sch.linkage( x, method = 'centroid' ) ) }" )
print( f"len( x ) = { len( x ) }" )
x.shape
```

```python
#linkage() is itself a algorithm for Hierachical Clusturing. It help in finding the Euclidean Distance between the points and
#forms the cluster.

sch_list = sch.linkage(x, method = 'ward' )
sch_list

#ward is the type of method used as a argument inside linkage() which helps in reducing the variance among the Clusters.
```

```python
len(sch_list)
```

```python
dg = sch.dendrogram( sch_list )
plt.xlabel( 'Euclidean Distance' )
plt.ylabel( 'Customers' )
plt.title( 'Dendogram' )
plt.show()
```

```python
#For Your Refrence

#Here in following codes, We can verify that icoord is the dimension of each sub - Traingle Cluster in Dendogram.

not_same_length = 0
not_same_breadth = 0

for i in dendogram['icoord']:
    if( i[0] != i[1] ):
        not_same_breadth += 1
    
    if( i[2] != i[3] ):
        not_same_length += 1
    
print( f"not_same_length = { not_same_length } " )
print( f"not_same_breadth = { not_same_breadth } " )
```

```python
#For Your Refrence

#Here we are checking the value of denndogram[ 'ivl' ] and denndogram[ 'leaves' ] whether both having same value or not.

#Only Diffrence is ivl key contains values in String Formate and Leaves key contains value in Integer Formate.

#Using Following Code, we can count how many values are not same ivl and leaves.

leaves_not_same = 0

for i in range( len( dendogram['ivl'] ) ):
    if( int( dendogram['ivl'][i] ) != dendogram['leaves'][i] ):
        leaves_not_same += 1
    
print( f"leaves_not_same = { leaves_not_same } " )

#We can see now, all the values are same
```

```python
#For Your Refrence

cost_of_iem = 12000
```

```python
#For Your Refrence

count = 0

buyer_category = []

for i in range( len( data ) ):  #0, 1, .., 199
    
    if( data['Spendable'][i] >= cost_of_iem ):
        count = count + 1
        buyer_category.append( data['y'][i] )

print( f"No. of Persons who will buy the Product = { count }" )
print( f"Buyer Category = { buyer_category }" )
```

```python
#For Your Refrence

cat_zero = 0
cat_one = 0
cat_two = 0
cat_three = 0
cat_four = 0

for i in buyer_category:
    if( i == 0 ):
        cat_zero = cat_zero + 1
    
    if( i == 1 ):
        cat_one = cat_one + 1
        
    if( i == 2 ):
        cat_two = cat_two + 1
        
    if( i == 3 ):
        cat_three = cat_three + 1
        
    if( i == 4 ):
        cat_four = cat_four + 1

print( f"Category Zero = { cat_zero }" )
print( f"Category One = { cat_one }" )
print( f"Category Two = { cat_two }" )
print( f"Category Three = { cat_three }" )
print( f"Category Four = { cat_four }" )
```
